```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
```

```python
qr = QuantumRegister(4,'q')
cr = ClassicalRegister(2,'c')
qc = QuantumCircuit(qr,cr)
qc.draw('mpl')
```

```python
def barrier1():
    qc.barrier([qr[i] for i in range(4)])
    return qc
def cx(x,y):
    qc.cx(qr[x],qr[y])
    return qc
def ccx(x,y,z):
    qc.ccx(qr[x],qr[y],qr[z])
    return qc
```

```python
qc.x(qr[0])
qc.x(qr[1])
barrier1()
cx(0,1)
cx(1,2)
ccx(0,1,3)
barrier1()
qc.measure(qr[3],cr[1])
qc.measure(qr[2],cr[0])
qc.draw('mpl')
```
