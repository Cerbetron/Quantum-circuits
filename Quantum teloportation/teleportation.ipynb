```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
```

```python
qr = QuantumRegister(3,'q')
cr = ClassicalRegister(3,'c')
qc = QuantumCircuit(qr,cr)
qc.draw('mpl')
```

```python
def barrier1():
    qc.barrier([qr[i] for i in range(3)])
    return qc
def cx(x,y):
    qc.cx(qr[x],qr[y])
    return qc
```

```python
qc.x(qr[0])
barrier1()
qc.h(qr[1])
qc.cx(1,2)
qc.cx(0,1)
qc.h(qr[0])
qc.measure(qr[0],cr[0])
qc.measure(qr[1],cr[1])
qc.cx(1,2)
qc.cz(qr[2],qr[0])

qc.measure(qr[2],cr[2])
qc.draw('mpl')
```
